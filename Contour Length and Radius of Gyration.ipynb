## Contour length

#### Raw contour length
The length of a single base pair, $L_{single\space bp} = 34.0$ nm. The raw contour length is given by multiplying the length of a single base pair with the number of base pairs:

$$L_{raw} = N_{bp}\cdot L_{single\space bp}$$

### Freely-jointed chain model
In the freely-jointed chain (FJC) model, a polymer is seen as a chain of N stiff, rod-like segments, with no limitations in the bond angles. With this model, DNA is seen as a random-walk where the segments have equal probability to fluctuate in all directions. The segment length is called the Kuhn length, b, and is two times the persistence length:
$$b = 2l_p$$

The contour length, L, is the length at full extension and is given by the product of the Kuhn length and the number of Kuhn segments: 

$$L = b N$$

The number of Kuhn segments, $N$ for a raw DNA strand is then calculated as $N = L_{raw}/b$.

### Persistence length

We assume the persistence length of raw DNA to be 50 nm [Dorfman, 2013].


#### Ionic strength
The ionic strength, $I$, is calculated by summing the product of the concentration and squared charge of all ions in the solution:

$$I = \dfrac{1}{2} \sum_{i=1}^{n} c_i z_i^2$$

where the one half is added to include both anions and cations, $c_i$ is the molar concentration of the ion i (in M) and $z_i$ is the charge of the ion i. 

#### Persistence length increase with decreasing salt level
The persistence length, $l_p$ changes with the ionic strength as the following (according to Odijk−Skolnick−Fixman (OSF) theory (see Beyondseq, Dorfman)): 

$$l_p = l_p^{'} + \dfrac{0.0324 M}{ I} nm$$

where $ l_p^{'} \approx 50$  nm is the bare persistence length.

An alternative, later model is given by Dobrynin. The persistence length, $l_p$ changes with the ionic strength as the following (according to Dobrynin (see Beyondseq, Dorfman)): 

$$l_p = 46.1 + \dfrac{1.9195 M}{ \sqrt{I}} nm$$

In this work, however, OSF-theory is used to model the length-increase by ionic strength.

Reisner et al., (2007) showed how the length of DNA-strands doubled when going a medium with high salt to low salt. They used the OSF-theory to explain their results [[Reisner, 2007]](https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.99.058302).

The contour length is increased with 0.51±0.14 nm for every incorporated dye molecule [Günther 2010]. For a staining ratio of 1 dye molecule per 10 base pairs, we get a 15% increase in the contour length. For a ratio of 1:4 we get a 37.5% increase.

$$L_{1:4} = 1.38\cdot L_{raw}$$

## Radius of Gyration

### Ideal Chain Radius Calculation

The worm-like chain (WLC) model (also called the Kratky-Porod model) describes the DNA dynamics more more accurately compared to the FJC-model by defining the polymer as semi-flexible. It means that the chain is stiff at the length scale of a monomer but flexible at the length scale of the entire polymer. The angle of a single segment affects the angle of its neighbour compared to FJC where the angles of segments are completely unrelated. The WLC model can be said to describe a polymer as a continuously flexible isotropic rod. The end-end distance of the polymer, $R_e$, is a measure of the size of the coiled-up polymer. The WLC model describes $R_e$ for long chains ($L≫L_p$) with [Rubinstein, 2003]:

$$R_e=b \cdot \sqrt{N} = b \cdot \sqrt{L/b}$$

Another way to characterize the size of a polymer is with the radius of gyration, $R_G$. The radius of gyration of a polymer is the root mean square distance from its center of mass. For an ideal chain, it is given by [Rubinstein, 2003]:

$$R_G^{\theta}=\dfrac{R_e}{\sqrt{6}}=b\sqrt{\dfrac{N}{6}}$$

$$ => $$

$$R_G^{\theta} \propto \sqrt{L}$$

$$R_G^{\theta} \propto l_p\sqrt{L}$$

### Real (non-ideal) chains
We have worked exclusively with good solvents (aqueous buffers). The ideal chain
models, described above, make assumptions that are not valid for good solvents.
They assume that the segments do not occupy any volume and they do not sterically
hinder other segments from passing through them, which they do. The same volume
can only be occupied once. This volume is then excluded for the rest of the chain to
occupy. The ideal models also ignore electrostatic interactions. In reality, the
negative-charged monomers repulse each other electrostatically, increasing the size
of the coil. To take these effects into account, the effective width, $𝑤_{𝑒𝑓𝑓}$, is used. It
includes both the steric and electrostatic interactions and is much larger than the
bare DNA width.

With real chains we then take the following into consideration:
- solvent quality (usually a "good" solvent when using aqeuous solutions, not a $\theta$-solvent which gives ideal conditions)
- excluded volume effect
- electrostatic interactions


Flory provided a scaling law where the excluded volume interactions are taken into consideration:

$R_g \propto N^{\nu}$

where $\nu  = 0.5877$ [Madras, 1995, see lic].

#### Measuring the radius experimentally
In contrast to $R_e$, $R_g$ can be easily measured experimentally. It is often calculated
with the Einstein relation by measuring the Brownian motion and acquiring the diffusion coefficients of DNA chains using fluorescent videography [Smith, 1996].

### Radius of gyration (high salt) as a function of solvent quality, z
The **radius of gyration** changes as a function of the solvent quality. The solvent quality varies highly within the same solvent with temperature. For example, for DNA 45 kbp in the buffer 1x TE and 0.5M NaCl, Pan et al. [Pan, 2014] deterimed the solvent quality to vary from 0 at 15$^\circ$C to 1.66 at 35$^\circ$C. This changes $\alpha_g(z)$ from 1 to 1.39, using the constants for z below.

$$R_g = R_g^{\theta}\alpha_g(z)$$

where $R_g^{\theta} = L/\sqrt{6N}$ is the radius of gyration during $\theta$-conditions and $\alpha_g(z)$ is the swelling ratio of the radius of gyration as a function of the chain interaction parameter, $z$.

The swelling ratio is given by:

$$\alpha_g(z) = (1+az+bz^2+cz^3)^{m/2}$$

where a = 9.528, b = 19.48, c = 14.92, and m = 0.1339 [Kumar, 2003]. These constants have been determined with Brownian Dynamics simulations [Pan, 2014].

#### Chain interaction parameter, z
The solvent quality (usually denoted z) for a $\theta$-solvent is 0 whereas it is infinite for a very good solvent. The $T_\theta$ for aqueous solutions (in excess Sodium salt)  for DNA was determined to be $T_\theta=14.7 \pm 0.5℃$ [Pan 2014]

$$z = k \left(1-\dfrac{T_\theta}{T}\right)\sqrt{M}$$
where $k$ is a constant depending on the chemistry. Pan et al. determined this to k = 0.0047 $\pm$ 0.0003 g/mol$^{-1/2}$ for the Tris-EDTA buffer solution they used (10 mM Tris, 1 mM EDTA [1x TE], 0.5 M NaCl and water) The 0.5M NaCl is to ensure excess salt to screen the charges.

If we assume we use the same buffer as Pan et al. (excess salt) [Pan, 2014, Thesis] for T = 22.9$^{\circ}$C and the same $k$ we get $z=0.7$ and $\alpha_g \approx 1.24$ for lambda-DNA. This is the same z as Pan et al. presents in their work.

#### Calculations of $\alpha_g$ and $z$ using the equations from the Prakash group

### Effect of ionic environment on the radius
It is important to note that most experiments conducted on the polymer physics of DNA use excess salt to screen the electrostatic interactions. 

When the other effects are also included, the following expression for the end-end distance is given:

$$R_e \approx (w_{eff} l_p)^{1/5}(bN)^{\nu} = (w_{eff} l_p)^{1/5}L^{\nu}$$

where $w_{eff}$ is the effective width of the polymer. Reisner et al. described the $w_{eff}$ for strongly charged chains as [[Reisner, 2007]](https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.99.058302):

$$w_{eff} = \dfrac{1}{\kappa}[0.7704+log(\dfrac{\nu^2_{eff}}{2\epsilon\epsilon_0 k_B T \kappa})]$$

where $1/\kappa$ is the Debye length (also sometimes denoted $\lambda_D$), $\epsilon$ the dielectric constant of water, $\epsilon_0$ the permittivity of free space, $\nu_{eff}$ is an effective DNA line charge density, $k_B$ is Boltzman's constant and $T$ the temperature. Note that [[Hsieh, 2008 (Doyle group)]](http://web.mit.edu/doylegroup/pubs/Nanolett_hsieh_08.pdf) used a constant of $2\pi$ instead of $0.5$ inside the log-expression. If I do this, I can replicate their values of $w_{eff}$.

$\kappa$  in turn is defined as:

$$\kappa^2  = \dfrac{2000N_Ae^2I}{\epsilon_0\epsilon k_BT}$$

where $N_A$ is Avogadro's number, $e$ the electronic charge and $I$ the ionic strength of the buffer. The Debye length should be on the order of tens of nanometer for low ionic strengths (~1 mM) and Å for higher ionic strenth buffers (see calculated values at [[Montes, 2019]](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC6932854/)).

Iarko et al. calculated the effective width of DNA at 5 x TBE to be 4.6 nm. [[Iarko, 2015]](https://publications.lib.chalmers.se/records/fulltext/225267/local_225267.pdf). This value is the same as I end up with in the calculation below.

#### Calculation: Debye length and effective width

```python
importlib.reload(dn)
T=23 #Temperature in Celcius
I = 43.62*1e-3 #Ionic strength of 5xTE and 3% BME buffer[M]
debye_length = dn.calc_Debye_length(T+293.15,I) # [m]
kappa = 1/debye_length # [m^-1]
w_eff_5xTE_23C = dn.calc_w_eff(I, T_deg_C=T)
print('5x TE, 3% BME and T=23C')
print(f'\tT = {T:.1f} C\n\tI = {dn.I_5xTE_003BME*1e3:.1f} mM')
print(f'\tDebye length = {debye_length*1e9:.1f} nm\n\tw_eff = {w_eff_5xTE_23C*1e9:.1f} nm')
print('\n')
T=23 #Temperature in Celcius
I = 1.22*1e-3 #Ionic strength of 0.1xTE buffer[M]
debye_length = dn.calc_Debye_length(T+293.15,I) # [m]
kappa = 1/debye_length # [m^-1]
w_eff_01xTE_23C= dn.calc_w_eff(I, T_deg_C=T)
print('0.1x TE and T=23C')
print(f'\tT = {T:.1f} C\n\tI = {I*1e3:.1f} mM')
print(f'\tDebye length = {debye_length*1e9:.1f} nm\n\tw_eff = {w_eff_01xTE_23C*1e9:.1f} nm')
```

#### Calculation: Flory Radius
The Flory radius is calculated for lambda-DNA at low salt and T=23C

```python
FloryExp = 0.5877
def calc_Flory_radius(w_eff, l_p,b,N):
    R_e = (w_eff * l_p)**(1/5) * (b*N)**FloryExp
    return R_e

l_p = 50e-9 # Persistence Length [m]
b = 2*l_p # Kuhn Length [m]
L_singleBP = 3.40e-10  # Length of one base pair [m]
N_bp = 48502 #Number of base pairs, lambda-DNA
L_raw = L_singleBP * N_bp
N = L_raw/b
R_e_Flory_T23C = calc_Flory_radius(w_eff_5xTE_23C, l_p,b,N)
# R_g_Flory_T23C = R_e_Flory_T23C / np.sqrt(6) #Not valid as the conditions are not ideal?

R_e_ideal = b * np.sqrt(N)
R_e_FloryExp = ((b*N)**FloryExp) * np.sqrt(b)

print('For buffer 5x TE, 3% BME')
print(f'R_e_ideal = {R_e_ideal*1e6:.2f} um')
print(f'R_e_FloryExp = {R_e_FloryExp*1e6:.2f} um')
print(f'R_e_Flory_T23C = {R_e_Flory_T23C*1e6:.2f} um')
# print(f'R_g_Flory_T23C = {R_g_Flory_T23C*1e6:.2f} um')
```
